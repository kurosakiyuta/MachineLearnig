最尤推定料の方程式

```
size = 1000
def max_lilelifood_estimation (size):
  a = np.random.randn(size)
  expectedvalue = sum(a)/size
  dispersion = ( sum( [(a[i] - expectedvalue)**2 for i in range (size)] ))/ size
  return dispersion
print(max_lilelifood_estimation (size))
```
